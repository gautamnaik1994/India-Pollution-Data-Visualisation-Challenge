```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel('All_Cities.xlsx',sheet_name="Sheet1")
```

```python
df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(15,8))
sns.lineplot(x="year" , y="SO2",data=df,label="SO2",palette="pastel")
sns.lineplot(x="year" , y="NO2",data=df,label="NO2",palette="pastel")
sns.lineplot(x="year" , y="PM10",data=df,label="PM10",palette="pastel")
plt.legend()

plt.style.use("dark_background")
plt.title("Pollution Over Time")
plt.ylabel("Polution")
plt.xlabel("Time")
```

```python
by_states=df.groupby("State")
```

```python
df_bystates=by_states.mean()
```

```python
# by_states['State'] = by_states.index
```

```python
df_bystates.reset_index(level=0,inplace=True)
```

```python
df_bystates
```

```python
temp=df_bystates.filter(["State","NO2","SO2","PM10"], axis=1)
tidy = temp.melt(id_vars='State').rename(columns=str.title)
# tidy.sort_values('Value').head(50)
```

```python
plt.figure(figsize=(15,30))
sns.barplot(y="State",x="Value",hue='Variable', data=tidy)
```

```python
sns.barplot(x="SO2",y="State", data=df_bystates.sort_values("SO2").tail(10), palette="Reds") 
```

```python
sns.barplot(x="NO2",y="State", data=df_bystates.sort_values("NO2").tail(10), palette="Reds") 
```

```python
sns.barplot(x="PM10",y="State", data=df_bystates.sort_values("PM10").tail(10), palette="Reds") 
```
